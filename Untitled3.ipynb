```python
# Рассмотрим матрицу корреляции для быстрого понимания какие признаки влияют на стоимость недвижимости. 
# На основе построенной матрицы по всем признакам таблицы, за исключением даты продажи, можно сделать вывод, что прямую связь с ценой недвижимости
# имеют следующие признаки:
# 1. Жилая площадь
# 2. Качество конструкции и дизайна
# 3. Кол-во ванных компнат
# 4. Жилая площадь 15 соседей.
import numpy as np
data=df.drop(columns=['date'])
corr_matrix = data.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');

```

```python
# Построим графики признаков с прямой зависимостью для наглядности.
# 1. Зависимость цены от жилой площади недвижимости
# Действительно чем больше площадь дома, тем выше его цена. Однако основные продажи происходят в сегменте недорогой до 1 млн.$

sns.jointplot(x=data['sqft_living'], y=data['price'], kind='reg',color='#548B54');

plt.xlabel('жилая площадь, кв.фут') 
plt.ylabel('стоимость недвижимости') 
```

```python
# Проверим насколько выгоднее продавать недорогую недвижимость. Узнаем сколько было получено за проданные дома свыше одного миллиона, 
# и сравним этот показатель с суммой продаж до одного миллиона долларов
#  Можно сделать вывод, что выгоднее продавать недвижимость обычным гражданам, заработать можно гораздо больше, чем продавая дорогую нед
data.loc[data['price'] <1000000.0, 'Cat_price']='До 1млн$' 
data.loc[data['price'] >1000000.0, 'Cat_price']='Свыше 1млн$'
plt.figure(figsize=(6, 3))
data.groupby('Cat_price')['price'].sum().plot(kind='bar',color='#548B54', edgecolor='black')
plt.xlabel('Категория стоимости недвижимости') 
```

```python
# 2. Качество конструкции и дизайна
# Можно сделать вывод, что чем выше качество недвижимости, тем дороже его стоимость
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['grade'], y=df['price'], showfliers=False, color='#548B54')

plt.xlabel('категории качества конструкции и дизайна')
plt.ylabel('Стоимость недвижимости')
plt.title('Влияние на стоимость качества недвижимости');
```

```python
# 3. Кол-во ванных компнат
# Здесь сложно сделать определенный вывод. Да, действительно самые дорогие - это самые высокие дома, однако популяроность все-таки относится
# к категории высоты дома от 1000 до 4000
# Чем больше ванных комнат, тем дороже недвижимость

plt.figure(figsize=(16, 8))

sns.boxplot(x=df['bedrooms'], y=df['price'], showfliers=False, color='#548B54')

plt.xlabel('Количество ванных комнат')
plt.ylabel('Стоимость недвижимости')
plt.title('Влияние на стоимость количества ванных комнат');
```

```python
# 4. Жилая площадь 15 соседей.
# Важен район, чем больше дома соседей, тем выше стоимость недвижимости 

sns.jointplot(x=data['sqft_living15'], y=data['price'], kind='reg',color='#548B54');

plt.xlabel('жилая площадь 15 соседей, кв.фут') 
plt.ylabel('стоимость недвижимости') 

```
